```python
#create a 5*5 array of random integers
#find mean of values row and column wise
#replace all even numbers with -1
#sort the array row and column wise
#find the row and column wise sum seperately
#count the number of -ve's in each column

import numpy as np
arr = np.random.randint(1,91,(5,5))
print(arr)

print("\n")
print("row wise mean : ")
print(np.mean(arr,axis=1))

print("\n")
print("column wise mean : ")
print(np.mean(arr,axis=0))

modified_arr = np.where(arr%2==0,-1,arr) #arr[arr % 2 == 0] = -1 this will not create new var
print("\n arr after replace evens with -1 :")
print(modified_arr)
print("\n")

print("row wise sorted arr :")
y = np.sort(arr, axis=1)
print(y)


print("\n")
x = np.sort(arr, axis=0)
print("column wise sorted arr :")
print(x)

print("\n")
print("row wise sum of arr :")
print(np.sum(arr,axis=1))

print("\n")
print("column wise sum of arr :")
print(np.sum(arr,axis=0))

print("\n")
print("number of -1 in each column")
print(np.sum(np.where(modified_arr<0,1,0),axis=0))
# print(np.sum(arr<0,axis=0))


```

```python
#question 2
#inverntory control system
#task:
#create 5*5 array(stock levels) range - 20 to 80
#create another 5*5 array (demand)
#compute:
#remaining_stock = stock - demand
#replace:
#negative values -> 0 (out of stock)
#find:
#total shotage
#products with zero stocks
#stock each based on remaining stock

import numpy as np
stock = np.random.randint(20,80,(5,5))
demand = np.random.randint(1,90,(5,5))
print('stock levels: ')
print(stock)
print("\n")

print('demand: ')
print(demand)
print("\n")
remaining_stock = stock - demand
print("remaing stock : ")
print(remaining_stock)
print("\n")
print('remaing after replacement:')
remaining_stock = np.where(remaining_stock<0,0,remaining_stock)
print(remaining_stock)

print("\n")
zero_stock=np.sum(np.where(remaining_stock==0,1,0),axis=0)
print("zero stocks:")
print(zero_stock)
```
